```python
# Web Scraping Quotes and Saving to CSV
```

```python
import requests
from bs4 import BeautifulSoup
import csv
```

```python
# website URL

URL = "https://quotes.toscrape.com/"
```

```python
# sending request to website

try:
    response = requests.get(URL)

    # checking if request was successful
    if response.status_code == 200:

        # parsing HTML content
        soup = BeautifulSoup(response.text, "html.parser")

        # finding all quotes
        quotes = soup.find_all("span", class_="text")

        # creating CSV file
        with open("Level2/quotes(t1).csv", "w", newline="", encoding="utf-8") as file:
            writer = csv.writer(file)

            # writing header
            writer.writerow(["Quote"])

            # writing quote data
            for quote in quotes:
                writer.writerow([quote.text])

        print("Data scraped and saved into quotes(t1).csv successfully!")

    else:
        print("Failed to retrieve webpage.")

except requests.exceptions.RequestException:
    print("Error while connecting to the website.")
```
