```python
# Bitcoin Price Tracker Using API
```

```python
import requests

# API URL
URL = "https://api.coingecko.com/api/v3/simple/price?ids=bitcoin&vs_currencies=inr"
```

```python
# sending request
try:
    response = requests.get(URL)

    # checking response status
    if response.status_code == 200:

        # converting response into JSON
        data = response.json()

        # extracting bitcoin price in INR
        bitcoin_price = data["bitcoin"]["inr"]

        # displaying output
        print("Bitcoin Price Tracker")
        print("----------------------")
        print("Current Bitcoin Price (INR): ₹", bitcoin_price)

    else:
        print("Failed to fetch data from API.")

except requests.exceptions.RequestException:
    print("Error: Unable to connect to API.")

except KeyError:
    print("Error: Invalid response received.")
```
