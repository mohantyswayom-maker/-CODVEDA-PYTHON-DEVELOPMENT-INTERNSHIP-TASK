```python
# function for encryption

def encrypt(text, shift):

    encrypted_text = ""

    for char in text:

        # checking if character is alphabet
        if char.isalpha():

            # for lowercase letters
            if char.islower():
                encrypted_text += chr((ord(char) - 97 + shift) % 26 + 97)

            # for uppercase letters
            elif char.isupper():
                encrypted_text += chr((ord(char) - 65 + shift) % 26 + 65)

        else:
            # keeping spaces and symbols unchanged
            encrypted_text += char

    return encrypted_text
```

```python
# function for decryption

def decrypt(text, shift):

    return encrypt(text, -shift)
```

```python
# main menu

print("1. Encrypt File")
print("2. Decrypt File")

choice = input("Enter your choice (1/2): ")

filename = input("Enter file name: ")

# shift value for Caesar cipher
shift = 3
```

```python
# encryption process

if choice == '1':

    try:

        # opening original file
        with open(filename, "r") as file:
            content = file.read()

        # encrypting content
        encrypted_content = encrypt(content, shift)

        # creating new encrypted file
        new_file = "encrypted_" + filename

        with open(new_file, "w") as file:
            file.write(encrypted_content)

        print("File encrypted successfully!")
        print("Encrypted file:", new_file)

    except FileNotFoundError:
        print("Error: File not found.")

# decryption process

elif choice == '2':

    try:

        # opening encrypted file
        with open(filename, "r") as file:
            content = file.read()

        # decrypting content
        decrypted_content = decrypt(content, shift)

        # creating decrypted file
        new_file = "decrypted_" + filename

        with open(new_file, "w") as file:
            file.write(decrypted_content)

        print("File decrypted successfully!")
        print("Decrypted file:", new_file)

    except FileNotFoundError:
        print("Error: File not found.")

# invalid option

else:
    print("Invalid choice.")
```
